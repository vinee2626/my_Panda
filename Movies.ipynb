## Movies data on different platforms

```python
import pandas as pd 
```

### Reading data and creating a dataframe

```python
df = pd.read_csv('MoviesOnStreamingPlatforms_updated.csv', index_col=0)
df.reset_index(drop=True)
df.head(5)
```

### Cleaning data

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(how='any', inplace=True)
df
```

```python
df.info()
```

## QUESTION 1: Which Comedy movies are available on Netflix?

```python
import re
new_df = df.loc[df['Netflix'] == 1 ]
col = list(df.columns)
new_df = new_df[col[0:6] + col[11:16]]
new_df = new_df[new_df['Genres'].str.contains('Comedy',flags = re.I, regex =True)]
new_df
```

```python
new_df.to_csv('Netflix_movies.csv', index=False)
```

## QUESTION 2: Which genres are watched the most by the customers ? 

```python
genres = df.groupby(['Genres']).count().sort_values( ascending=False)
genres.head(10)
#these are the top 10 preffered genres of movies on all platforms
```

## QUESTION 3: Which are the Adult movies? And among them which section contributes to the most views ?

#### Adding a new column for the type of Audience and mentioning adults and kids movies 

```python
df['Audience'] = ''
df.loc[df['Age'] == '18+', 'Audience'] = 'Adult'
df.loc[df['Age'] == '7+', 'Audience'] = 'Kids'
df.loc[df['Age'] == '13+', 'Audience'] = 'Under Parental Guidance'
df.loc[df['Age'] == 'all', 'Audience'] = 'Universal'
df.loc[df['Age'] == '16+', 'Audience'] = 'Teens' 
df
```

### Plotting a graph to identify the most active age group

```python
import matplotlib.pyplot as plt
audiences= [audience for audience ,df in df.groupby('Audience')]

plt.plot(audiences, df.groupby(['Audience']).count())
plt.xticks(audiences, rotation= 'vertical')
plt.xlabel('Types of audience',size=10, color = 'r')
plt.ylabel('Number of views', size=10, color = 'r')
plt.show()
```

## QUESTION 4: People prefer which language movies the most? And why do you think they it is voted the most? 

```python
df['Primary language'] =df['Language'].str.split(',',expand =True )[0]
df.groupby(['Primary language']).count()['ID'].sort_values(ascending= False)
#English is the most preffered choice of language for movie on platforms 
```
